menggunakan **parse dates** untuk memahami lebih jelas format waktu yang digunakan dalam **time_update**

```python
belanja = pd.read_csv('data_input/online_bl.csv', parse_dates=['time_update'])
belanja.head()
```

mencari unique value dalam file online_bl  

```python
belanja.describe(include='all').loc['unique']
```

dilihat dari hasil di atas **sub_category** dan **scale** memiliki unique value dengan masing-masing jumlah terdapat 3 dan 4. Berikutnya dapat melihat isi dari unique value tersebut

```python
belanja['sub_category'].unique()
```

```python
belanja['sub_category'].value_counts()
```

```python
belanja['scale'].unique()
```

```python
belanja['scale'].value_counts()
```

selanjutnya bisa mencari insight dengan membandingkan dari sub_category dan scale untuk mengetahui distribusi dari nilai yang ada dengan bantuan crosstab

```python
pd.crosstab(
    index=belanja['scale'],
    columns=belanja['sub_category']
).sort_values(by='detergent', ascending=False)
```

bisa juga mencari insight rata-rata penjualan dengan membandingkan dari kolom lainnya seperti price_original dan juga waktu yang ada

```python
belanja['bulan']=belanja['time_update'].dt.month_name()
pd.crosstab(
    index=belanja['bulan'],
    columns=belanja['scale'],
    values=belanja['price_original'],
    aggfunc='mean'
).sort_values(by='1.8 kg', ascending=True)
```

berdasarkan total (sum), ascending berdasarkan berat 1.8 kg

```python
belanja['bulan']=belanja['time_update'].dt.month_name()
pd.crosstab(
    index=belanja['bulan'],
    columns=belanja['scale'],
    values=belanja['price_original'],
    aggfunc='sum'
).sort_values(by='1.8 kg', ascending=True)
```

mencoba dataset lain dari **techcruch**

```python
tech = pd.read_csv('data_input/techcrunch.csv')
tech.head()
```

memastikan formatting yang ada dari tiap kolom

```python
tech.info()
```

```python
pd.crosstab(
    index=tech['state'],
    columns=tech['category'],
    values=tech['raisedAmt'],
    aggfunc='median'
).sort_values(by=['mobile','cleantech'],ascending=[False,False])
```

```python
pd.crosstab(
    index=tech['state'],
    columns=tech['category'],
    values=tech['raisedAmt'],
    aggfunc='median'
).sort_values(by=['consulting','biotech'],ascending=[False,False])
```

```python
tech = pd.read_csv('data_input/techcrunch.csv', parse_dates=['fundedDate'])
tech.head()
```

```python
tech.describe(include='all').loc['unique']
```

mencari trending friendster terbanyak dari kolom permalink berdasarkan waktu

```python
time = tech[tech['permalink'] == 'friendster'].copy()
time['Waktu'] = time['fundedDate'].dt.to_period('M')
pd.crosstab(
    index=time['Waktu'],
    columns=time['permalink'],
    values=time['raisedAmt'],
    aggfunc='max',
).sort_values(by='friendster',ascending=False)
```

```python
city = tech[tech['city'] == 'San Francisco'].copy()
pd.crosstab(
    index=city['permalink'],
    columns=city['city'],
    values=city['raisedAmt'],
    aggfunc='sum'
).sort_values(by='San Francisco', ascending= False)
```
